```python
# let's create a data frame for the train performance evaluation results set.
import pandas as pd
columnNames = ['correctCount', 'accuracy', 'meanSquareError', 'crossEntropy', 'lHC']
df_train = pd.DataFrame(resTrain4[0], columns=[columnNames])
df_train.head()
```

```python
# let's create a data frame for the test performance evaluation results set.
import pandas as pd
columnNames = ['correctCount', 'accuracy', 'meanSquareError', 'crossEntropy', 'lHC']
df_test = pd.DataFrame(resTrain4[1], columns=[columnNames])
df_test.head()
```

### A. plots that compares the error curves of the three cost functions for the training.

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
line_mse = plt.plot(range(0,50), df_train['meanSquareError'], 'r')
line_crossEntropy = plt.plot(range(0, 50), df_train['crossEntropy'], 'g')
line_lHC = plt.plot(range(0, 50), df_train['lHC'], 'b')
plt.title('Error by Cost Function (training)')
plt.ylabel('Error')
plt.xlabel('Epoch')
plt.legend(['MSE', 'CrossEntropy', 'LogLikelihood'])
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 7))
line_mse = plt.plot(range(0,50), df_test['meanSquareError'], 'r')
line_crossEntropy = plt.plot(range(0, 50), df_test['crossEntropy'], 'g')
line_lHC = plt.plot(range(0, 50), df_test['lHC'], 'b')
plt.title('Error by Cost Function (test)')
plt.ylabel('Error')
plt.xlabel('Epoch')
plt.legend(['MSE', 'CrossEntropy', 'LogLikelihood'])
plt.show()
```

### B. Three plots that shows the error for training vs. test

```python
plt.figure(figsize=(10, 7))
line_mse = plt.plot(range(0,50), df_train['meanSquareError'], 'r')
line_crossEntropy = plt.plot(range(0,50), df_test['meanSquareError'], '#cdc50a')
plt.title('Train vs. Test: MSE')
plt.ylabel('Error')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'])
plt.show()
```

```python
plt.figure(figsize=(10, 7))
line_mse = plt.plot(range(0,50), df_train['crossEntropy'], 'g')
line_crossEntropy = plt.plot(range(0,50), df_test['crossEntropy'], '#cdc50a')
plt.title('Train vs. Test: CrossEntropy')
plt.ylabel('Error')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'])
plt.show()
```

```python
plt.figure(figsize=(10, 7))
line_mse = plt.plot(range(0,50), df_train['lHC'], 'b')
line_crossEntropy = plt.plot(range(0,50), df_test['lHC'], '#cdc50a')
plt.title('Train vs. Test: logLikelihood')
plt.ylabel('Error')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'])
plt.show()
```
